```python
from transformers import pipeline

# Load pre-trained sentiment analysis model
sentiment_pipeline = pipeline("sentiment-analysis")

# Load pre-trained emotion detection model
emotion_pipeline = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)

def analyze_text_sentiment(text):
    result = sentiment_pipeline(text)[0]
    return result['label'], result['score']

def analyze_text_emotion(text):
    result = emotion_pipeline(text)[0]
    emotions = {label['label']: label['score'] for label in result}
    return emotions
```

```python
from pydub import AudioSegment
import torch
import torchaudio

def analyze_voice_tone(audio_path):
    # Load audio file
    audio = AudioSegment.from_file(audio_path)

    waveform, sample_rate = torchaudio.load(audio_path)
    return "calm"
```

```python
import cv2
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image

model = models.resnet50(pretrained=True)
model.eval()

transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def analyze_facial_expression(image_path):
    # Load image
    image = Image.open(image_path)
    image = transform(image).unsqueeze(0)

    with torch.no_grad():
        output = model(image)

    return "happy"
```

```python
def generate_therapeutic_response(emotion):
    responses = {
        "happy": "That's wonderful to hear! What's making you feel so good today?",
        "sad": "I'm here for you. Sometimes it helps to talk about what's bothering you.",
        "angry": "I understand that you're frustrated. Let's take a moment to calm down and figure this out together.",
        "calm": "You seem calm. How can I assist you today?",
        "neutral": "You seem neutral. What's on your mind?"
    }
    return responses.get(emotion, "I'm not sure how you're feeling. Can you tell me more?")
```

```python
def empathai_chatbot(text=None, audio_path=None, image_path=None):
    emotion = "neutral"

    if text:
        sentiment, _ = analyze_text_sentiment(text)
        emotions = analyze_text_emotion(text)
        dominant_emotion = max(emotions, key=emotions.get)
        emotion = dominant_emotion.lower()

    if audio_path:
        voice_emotion = analyze_voice_tone(audio_path)
        emotion = voice_emotion.lower()

    if image_path:
        facial_emotion = analyze_facial_expression(image_path)
        emotion = facial_emotion.lower()

    response = generate_therapeutic_response(emotion)
    return response

text = "I'm feeling soo good today."
audio_path = "/content/audio.wav"
image_path = "/content/happy.jpg"

response = empathai_chatbot(text=text, audio_path=audio_path, image_path=image_path)
print(response)
```
