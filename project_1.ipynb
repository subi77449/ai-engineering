```python
import numpy as np

# Dataset
data = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Calculate overall median
median = np.median(data)

# Split the data into lower and upper halves
lower_half = [x for x in data if x < median]
upper_half = [x for x in data if x > median]

# Calculate the medians of the halves
lower_median = np.median(lower_half)
upper_median = np.median(upper_half)

print(f"Overall Median: {median}")
print(f"Lower Median: {lower_median}")
print(f"Upper Median: {upper_median}")
```
