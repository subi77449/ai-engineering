```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/old.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, mask = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)

restored = cv2.inpaint(image, mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Damaged Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title("Damage Mask")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(restored, cv2.COLOR_BGR2RGB))
plt.title("Restored Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
