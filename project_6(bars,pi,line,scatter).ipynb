```python
data_subset = data.iloc[:10, [0, -1]].dropna()

countries = data_subset.iloc[:, 0]
gdp_values = data_subset.iloc[:, 1]
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=countries, y=gdp_values, palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars, fmt="%.2f")

    plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('GDP (2023)')
plt.title('GDP of First 10 Countries (2023)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(gdp_values, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('GDP Distribution of First 10 Countries (2023)')
plt.show()
```

```python
df_subset = data.iloc[:10, 4:14]
years = df_subset.columns[df_subset.columns != 'Country Name'].astype(int)
```

```python
plt.figure(figsize=(10, 6))
for i in range(len(df_subset)):
    plt.plot(years, df_subset.iloc[i, df_subset.columns != 'Country Name'], marker='o', label=data.iloc[i, 0])
plt.xlabel('Year')
plt.ylabel('GDP (US$)')
plt.title('GDP Trend for First 10 Countries')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
countries = data.iloc[:10, 4:15]
for i, (row, country) in enumerate(zip(df_subset.values, countries)):
    plt.scatter(range(len(row)), row, label=country, marker='o')

plt.xlabel("Column Index")
plt.ylabel("Values")
plt.title("Scatter Plot of First 10 Columns (Rows 4 to 15)")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
