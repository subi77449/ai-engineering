```python
import speech_recognition as sr

def recognize_speech_from_mic(recognizer, microphone):
    with microphone as source:
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    response = {
        "success": True,
        "error": None,
        "transcription": None
    }

    try:
        response["transcription"] = recognizer.recognize_google(audio)
    except sr.RequestError:
        response["success"] = False
        response["error"] = "API unavailable"
    except sr.UnknownValueError:
        response["error"] = "Unable to recognize speech"

    return response
```

```python
from transformers import pipeline

nlp_pipeline = pipeline("text-generation")

def process_text(text):

    if "book" in text.lower():
        return "book"
    elif "cancel" in text.lower():
        return "cancel"
    else:
        return "unknown"
```

```python
bookings = {}

def book_appointment(name, date, time):
    if date not in bookings:
        bookings[date] = {}
    if time not in bookings[date]:
        bookings[date][time] = name
        return True
    return False

def cancel_appointment(name, date, time):
    if date in bookings and time in bookings[date] and bookings[date][time] == name:
        del bookings[date][time]
        return True
    return False
```

```python
from gtts import gTTS
import os

def text_to_speech(text):
    tts = gTTS(text=text, lang='en')
    tts.save("response.mp3")
    os.system("mpg321 response.mp3")
```

```python
def voice_bot_agent():
    recognizer = sr.Recognizer()

    microphone_device_index = None

    try:
        microphone = sr.Microphone(device_index=microphone_device_index)

        print("Listening...")
        speech = recognize_speech_from_mic(recognizer, microphone)

        if speech["success"]:
            print("You said: {}".format(speech["transcription"]))
            intent = process_text(speech["transcription"])

            if intent == "book":
                name = "John Doe"
                date = "2025-05-20"
                time = "10:00"
                if book_appointment(name, date, time):
                    response = "Appointment booked for {} on {} at {}.".format(name, date, time)
                else:
                    response = "Sorry, the slot is already booked."
            elif intent == "cancel":
                name = "John Doe"
                date = "2025-05-20"
                time = "10:00"
                if cancel_appointment(name, date, time):
                    response = "Appointment canceled for {} on {} at {}.".format(name, date, time)
                else:
                    response = "No appointment found to cancel."
            else:
                response = "I didn't understand that. Can you please repeat?"

            text_to_speech(response)
            print(response)
        else:
            if speech["error"]:
                print("ERROR: {}".format(speech["error"]))

    except OSError as e:
        print(f"Error initializing microphone: {e}")
        print("Please ensure a microphone is connected and recognized by the system.")

voice_bot_agent()
```
