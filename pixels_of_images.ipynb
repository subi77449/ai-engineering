```python
from IPython.display import display
from PIL import Image

im = Image.open('/content/imagemasking.jpg', 'r')

# Call the display function with the image object to display it
display(im)
```

```python
pix_val = list(im.getdata())
```

```python
display(pix_val)
```

```python
with open('pixel_values.txt', 'w') as f:
    for pixel in pix_val:
        f.write(str(pixel) + '\n')  # Convert pixel to string and add newline
print("Pixel values saved to pixel_values.txt")
```

```python
import numpy as np

image = np.random.randint(low=0, high=255, size=(10, 10))

import matplotlib.pyplot as plt # Make sure to import pyplot

fig, ax = plt.subplots()
ax.imshow(image, cmap='gray')

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        ax.text(j, i, str(image[i, j]), color='r', ha='center', va='center')
```
